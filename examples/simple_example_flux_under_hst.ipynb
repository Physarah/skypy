```python
from skypy.skymods.earthshine import CeresSkyModel
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = datetime.datetime(year = 2007, month = 3, day = 13, hour = 14, minute = 20, second = 40)
end = datetime.datetime(year = 2007, month = 3, day = 13, hour = 14, minute = 30, second = 40)
ceres_sky = CeresSkyModel(start, end)
```

```python
ceres_data = ceres_sky.download_allsky_data(saveloc = '/Users/physarah/Desktop', 
                               data_product = "init_all_toa_sw_up")
plt.imshow(ceres_data, cmap='viridis', interpolation='nearest')
plt.ylim(175,0)
plt.show()
```
